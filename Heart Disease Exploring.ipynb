```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import seaborn as sns


```

```python
# study data files 

Heart_path = "data/heart.csv"

#read the data files 

data = pd.read_csv(Heart_path)

data.head()
```

```python
type(data)
```

```python
data.shape
```

```python
y= data["HeartDisease"]

plt.legend(["Does not have Disease", "Has Heart Disease"])

sns.countplot(y)

hd_temp = data.HeartDisease.value_counts()


print(hd_temp)
```

```python
print("Percentage of patients without Heart Disease:" +str(round(hd_temp[0]*100/918,2)))
print("Percentage of patients with Heart Disease: " + str(round (hd_temp[1]*100/918,2)))
```

```python
data["Sex"].unique()
```

```python
sns.barplot(data["Sex"],y)
```

```python
pd.crosstab(data.Age,data.HeartDisease).plot(kind="bar",figsize=(20,6))
plt.title('Heart Disease Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('heartDiseaseAndAges.png')
plt.show()
```

```python
pd.crosstab(data.Sex,data.HeartDisease).plot(kind="bar",figsize=(20,10),color=['blue','#AA1111' ])
plt.title('Heart Disease Frequency by Gender')
plt.xlabel('Sex (F = Female, M = Male)')
plt.xticks(rotation=0)
plt.legend(["Does not have Disease", "Has Heart Disease"])
plt.ylabel('Frequency')
plt.show()
```

```python
pd.crosstab(data.FastingBS,data.HeartDisease).plot(kind="bar",figsize=(20,10),color=['#4286f4','#f49242'])
plt.title("Heart disease according to FBS")
plt.xlabel('FBS- (Fasting Blood Sugar > 120 mg/dl) (1 = true; 0 = false)')
plt.xticks(rotation=90)
plt.legend(["Don't Have Disease", "Have Disease"])
plt.ylabel('Disease or not')
plt.show()
```
